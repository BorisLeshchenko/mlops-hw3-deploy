```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib
import pathlib
```

```python
# Загрузка и разбиение
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)
```

```python
# Обучение и оценка
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

preds = model.predict(X_test)
acc = accuracy_score(y_test, preds)
acc
```

```python
import pathlib
import joblib

# текущая папка = notebooks
notebooks_dir = pathlib.Path().resolve()
project_root = notebooks_dir.parent  # поднимаемся на уровень вверх

app_dir = project_root / "app"
app_dir.mkdir(exist_ok=True)

model_path = app_dir / "model.pkl"
joblib.dump(model, model_path)

```

```python
# Создаем новую версию модели
from sklearn.ensemble import RandomForestClassifier
import pathlib, joblib

data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=0
)

model_v2 = RandomForestClassifier(n_estimators=50, random_state=0)
model_v2.fit(X_train, y_train)

acc_v2 = accuracy_score(y_test, model_v2.predict(X_test))
acc_v2
```

```python
# сохранение модели
project_root = pathlib.Path().resolve().parent
model_path_v2 = project_root / "app" / "model_v1_1.pkl"
joblib.dump(model_v2, model_path_v2)
model_path_v2
```
